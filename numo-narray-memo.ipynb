# numo/narray まとめ

## はじめに実行

```ruby
require "gr/plot"
require "numo/narray"
```

## ベクトル

### Array からベクトル

```ruby
v = [3, 1, 2, 4]
Numo::Int32[*v]
```

```ruby
Numo::DFloat[*Array.new(10){|i| i / 10.0}]
```

### Range からベクトル

```ruby
Numo::Int32[1..10]
```

### 等間隔なベクトル

```ruby
Numo::Int32.linspace(-10, 10, 5)
```

### 零ベクトル

```ruby
Numo::Int32.zeros(10)
```

### 要素がすべて1のベクトル

```ruby
Numo::Int32.ones(10)
```

## 行列

### Array から行列

```ruby
m = [[1, 2], [3, 4]]
Numo::Int32[*m]
```

### 零行列の生成

```ruby
Numo::Int32.zeros(3, 3)
```

### 単位行列

```ruby
Numo::Int32.eye(3, 3)
```

## 計算

### 転置行列

```ruby
m = Numo::Int32[[1, 2],[3, 4]]
m.transpose
```

### 行列の積

```ruby
m = Numo::Int32[[1, 2], [3, 4]]
v = Numo::Int32[1, 2]
m.dot v
```

```ruby
m = Numo::Int32[[1, 1, -1], [-2, 0, 1], [0, 2, 1]]
Numo::DFloat.eye(3) / m
```

### フラットにする

```ruby
m = Numo::Int32[[1, 2], [3, 4]]
m.flatten
```

## 情報の取得

```ruby
e = Numo::Int32[[1, 2], [3, 4]]
p e.shape # 次元数
p e.size # 要素数
p e.rank # ランク
```
